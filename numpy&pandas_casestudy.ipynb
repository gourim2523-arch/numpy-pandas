```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv("IMDB-Movie-Data.csv")
```

```python
print(df.shape)
```

```python
print(df.head())
```

```python
print(df.columns)
```

```python
print(df.dtypes)
```

```python
print(df.isnull().sum())
```

```python
print("Duplicate Rows:", df.duplicated().sum())
```

```python
df = df.drop_duplicates()
```

```python
df.rename(columns={"Title": "Movie_Title"}, inplace=True)
print(df.columns)
```

```python
genres = df['Genre'].str.split(',').explode().unique()
print(genres)
```

```python
df['Genre_Count'] = df['Genre'].apply(lambda x: len(x.split(',')))
print(df[['Movie_Title', 'Genre_Count']].head())
```

```python

print("Missing values before filling:")
print(df[['Revenue (Millions)', 'Metascore']].isnull().sum())
df['Revenue (Millions)'] = df['Revenue (Millions)'].fillna(
    df['Revenue (Millions)'].mean()
)

```

```python
df['Metascore'] = df['Metascore'].fillna(
    df['Metascore'].mean()
)
print("\nMissing values after filling:")
print(df[['Revenue (Millions)', 'Metascore']].isnull().sum())
```

```python
print(df[['Revenue (Millions)', 'Metascore']].isnull().sum())
```

```python
top5 = df.nlargest(5, 'Rating')
print(top5[['Movie_Title', 'Rating']])
```

```python
highest = df.loc[df['Revenue (Millions)'].idxmax()]
print(highest[['Movie_Title', 'Revenue (Millions)']])
```

```python
lowest = df.loc[df['Revenue (Millions)'].idxmin()]
print(lowest[['Movie_Title', 'Revenue (Millions)']])
```

```python
avg_rating_director = df.groupby('Director')['Rating'].mean()
print(avg_rating_director)
```

```python
movies_per_year = df['Year'].value_counts().sort_index()
print(movies_per_year)
```

```python
result = df[
    (df['Rating'] > 8) &
    (df['Revenue (Millions)'] > 100)
]

print(result[['Movie_Title', 'Rating', 'Revenue (Millions)']])
```

```python
director_stats = df.groupby('Director').agg({
    'Rating':'mean',
    'Revenue (Millions)':'mean'
})

print(director_stats)
```

```python
year_stats = df.groupby('Year').agg({
    'Movie_Title':'count',
    'Rating':'mean'
})

year_stats.rename(columns={
    'Movie_Title':'Movie_Count',
    'Rating':'Avg_Rating'
}, inplace=True)

print(year_stats)
```

```python
df['Profitability'] = (
    df['Revenue (Millions)'] /
    df['Runtime (Minutes)']
)
```

```python
top10_profit = df.nlargest(10, 'Profitability')

print(top10_profit[
    ['Movie_Title',
     'Revenue (Millions)',
     'Runtime (Minutes)',
     'Profitability']
])
```

```python
def categorize_rating(rating):
    if rating < 6:
        return "Poor"
    elif rating < 7.5:
        return "Average"
    else:
        return "Excellent"

df['Rating_Category'] = df['Rating'].apply(categorize_rating)

print(df[['Movie_Title', 'Rating', 'Rating_Category']].head())
```

```python
def movie_score(row):
    return row['Rating'] * 10 + row['Revenue (Millions)']
```

```python
df['Movie_Score'] = df.apply(movie_score, axis=1)
```

```python
print(df[['Movie_Title',
          'Rating',
          'Revenue (Millions)',
          'Movie_Score']].head())
```

```python
top_score = df.nlargest(10, 'Movie_Score')

print(top_score[
    ['Movie_Title',
     'Rating',
     'Revenue (Millions)',
     'Movie_Score']
])
```
